### Trying different methods

##### Utility functions

```python
def show_image(img):
    cv2.imwrite('tmp/image.jpeg', img)
    subprocess.run(['xdg-open', 'tmp/image.jpeg'])
```

##### cv2

```python
# Load image
image = cv2.imread(sf)

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to preprocess the image
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

# Apply dilation to thicken the text
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
dilated = cv2.dilate(thresh, kernel, iterations=1)

# Extract text using pytesseract
text = pytesseract.image_to_string(dilated)

print(text)
```

##### cv2 Preprocessing: contrast, shrpening, noise removal

```python
def preprocess_image(image_path: str) -> np.ndarray:
    # read the image
    img = cv2.imread(image_path)
        
    # apply non-local means denoising
    denoised = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)

    # Convert image to grayscale
    gray = cv2.cvtColor(denoised, cv2.COLOR_BGR2GRAY)

    # increase the contrast
    alpha = 1.5  # contrast control (1.0-3.0)
    beta = 0    # brightness control (0-100)
    adjusted = cv2.convertScaleAbs(gray, alpha=alpha, beta=beta)

    # increase the sharpness
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    sharpened = cv2.filter2D(adjusted, -1, kernel)

    return sharpened
```

```python
# Pass the preprocessed image to pytesseract
text = pytesseract.image_to_string(preprocess_image(sf), lang='eng')

print(text)
```

##### Thresholding and denoise

```python
img = cv2.imread(sf, 0)
filtered = cv2.adaptiveThreshold(img.astype(np.uint8), 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 41,
                                    3)
kernel = np.ones((1, 1), np.uint8)
opening = cv2.morphologyEx(filtered, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
or_image = cv2.bitwise_or(img, closing)

# Pass the preprocessed image to pytesseract
text = pytesseract.image_to_string(or_image, lang='eng')

print(text)
```

##### Recognized Data

```python
img = preprocess_image(sf)
data = pytesseract.image_to_data(img, output_type='data.frame')
data
```

```python
# Get the bounding boxes from pytesseract
img = preprocess_image(sf)
d = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
n_boxes = len(d['level'])

# Draw the bounding boxes on the image
for i in range(n_boxes):
    if d['text'][i] != '':
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Get the text and draw it above the rectangle
        text = d['text'][i]
        cv2.putText(img, text, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

        text = str(d['block_num'][i])
        cv2.putText(img, text, (x, y-24), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

show_image(img)
```
